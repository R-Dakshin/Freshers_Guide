```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
URL = "https://chennai.vit.ac.in/freshers-induction-programme-2026/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

import requests
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

response = requests.get(
    URL,
    headers=headers,
    verify=False,
    timeout=30
)

print("Status Code:", response.status_code)

soup = BeautifulSoup(response.text, "html.parser")

rows = soup.find_all("tr")

departments = []

for row in rows:

    cols = row.find_all("td")

    if len(cols) < 6:
        continue

    department = cols[0].get_text(" ", strip=True)
    reporting_date = cols[1].get_text(" ", strip=True)
    orientation_date = cols[2].get_text(" ", strip=True)
    reporting_time = cols[3].get_text(" ", strip=True)
    venue = cols[4].get_text(" ", strip=True)

    link = cols[5].find("a")

    schedule_url = ""

    if link:
        schedule_url = urljoin(URL, link["href"])

    departments.append({
        "Department": department,
        "Reporting Date": reporting_date,
        "Orientation Date": orientation_date,
        "Reporting Time": reporting_time,
        "Venue": venue,
        "Schedule URL": schedule_url
    })

dept_df = pd.DataFrame(departments)

print("Departments Found:", len(dept_df))

dept_df.head()
```

```python
writer = pd.ExcelWriter(
    "Freshers_Induction_2026.xlsx",
    engine="openpyxl"
)

# Save department list
dept_df.to_excel(
    writer,
    sheet_name="Department List",
    index=False
)

print("="*60)

for i, row in dept_df.iterrows():

    department = row["Department"]
    url = row["Schedule URL"]

    print(f"{i+1}. {department}")

    if url == "":
        print("   No schedule URL")
        continue

    try:

        # Read all HTML tables
        html = requests.get(
          url,
          headers=headers,
          verify=False,
          timeout=30).text
        tables = pd.read_html(html)

        if len(tables) == 0:
            print("   No HTML table found")
            continue

        # Usually first table contains schedule
        schedule = tables[0]

        # Excel sheet names max 31 chars
        sheet_name = department[:31]

        schedule.to_excel(
            writer,
            sheet_name=sheet_name,
            index=False
        )

        print("   Saved")

    except Exception as e:
        print("   Error:", e)

writer.close()

print("\nExcel file created successfully!")
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

session = requests.Session()
session.verify = False

headers = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
all_data = []

for index, row in dept_df.iterrows():

    department = row["Department"]
    url = row["Schedule URL"]

    print(f"Processing: {department}")

    try:
        response = session.get(url, headers=headers, timeout=30)
        soup = BeautifulSoup(response.text, "html.parser")

        table = soup.find("table", class_="wp-acad-sched-table")

        if table is None:
            print("  No schedule table found.")
            continue

        tbody = table.find("tbody")

        for tr in tbody.find_all("tr"):

            cols = tr.find_all("td")

            if len(cols) != 5:
                continue

            all_data.append({
                "Department": department,
                "Date": cols[0].get_text(strip=True),
                "Day": cols[1].get_text(strip=True),
                "Time": cols[2].get_text(strip=True),
                "Details": cols[3].get_text(" ", strip=True),
                "Venue": cols[4].get_text(" ", strip=True)
            })

        print("  Done")

    except Exception as e:
        print("  Error:", e)
```

```python
schedule_df = pd.DataFrame(all_data)

schedule_df.to_excel(
    "Freshers_Induction_2026_Schedule.xlsx",
    index=False
)

print(schedule_df.head())
```

```python
import pandas as pd
import json

# Read the Excel generated earlier
df = pd.read_excel("Freshers_Induction_2026_Schedule.xlsx")

df.head()
```

```python
from datetime import datetime

programmes = []

# Group by programme
for programme, prog_df in df.groupby("Department"):

    schedule = []

    # Group by date
    for date, date_df in prog_df.groupby("Date"):

        # Convert DD.MM.YYYY -> YYYY-MM-DD
        try:
            formatted_date = datetime.strptime(str(date), "%d.%m.%Y").strftime("%Y-%m-%d")
        except:
            formatted_date = str(date)

        slots = []

        for slot_no, (_, row) in enumerate(date_df.iterrows(), start=1):

            slots.append({
                "slot": slot_no,
                "time": row["Time"],
                "activity": row["Details"]
            })

        schedule.append({
            "date": formatted_date,
            "slots": slots
        })

    programmes.append({
        "programme": programme,
        "schedule": schedule
    })
```

```python
with open("Freshers_Induction_2026.json", "w", encoding="utf-8") as f:
    json.dump(programmes, f, indent=2, ensure_ascii=False)

print("JSON Saved Successfully!")
```
